```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#Read the data in and remove any missing values
filename='C:/Users/dimpa/OneDrive - stevens.edu/CS-513/HW_04_NB/breast-cancer-wisconsin.csv'
nan_values=['?']
att_type={' Sample': 'str', 'Class':'category'}

df=pd.read_csv(filename,dtype=att_type,na_values=nan_values)
```

```python
print(df[df.isna().any(axis=1)])

print(df.shape)
print(df.isnull().sum())
print()
df.dropna(inplace=True) #Remove any row with a missing value in any of the columns.
df['F6'] = df['F6'].astype('int64')

print(f"shape : {df.shape}")
print(df.isnull().sum())
```

```python
print(df.dtypes)
df.head()
```

```python
# Splitting the dataset
attr = df.drop(columns = ['Class','Sample'])  # features
target = df['Class']  # target variable
X_train, X_test, Y_train,Y_test  = train_test_split(attr, target,test_size = 0.3, random_state = 22, shuffle = True)
Y_train.head()
Y_test.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
CART = DecisionTreeClassifier()

CART.fit(X_train, Y_train)
target_pred = CART.predict(X_test)
accuracy = accuracy_score(Y_test,target_pred ) 
print(f'Accuracy of model {accuracy}%')
print('')
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(50,30), dpi=250)
plot_tree(CART,fontsize=20,filled=True,feature_names=attr.columns.tolist());
```
