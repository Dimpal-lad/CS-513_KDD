```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Generate some random data for demonstration
np.random.seed(111)
data = [[2],[5],[9],[15],[16],[18],[25],[33],[33],[45]]

# Perform hierarchical clustering
linkage_matrix = linkage(data, method='single')

# Specify the number of clusters you want
num_clusters = 1

# Assign data points to clusters using fcluster
clusters = fcluster(linkage_matrix, t=num_clusters, criterion='maxclust')

# Plot the dendrogram with color-coded clusters
dendrogram(linkage_matrix, color_threshold=num_clusters, labels=np.arange(1, 11))

plt.title('Hierarchical Clustering Dendrogram with {} Clusters'.format(num_clusters))
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Display the assigned clusters for each data point
print("Assigned clusters:", clusters)
```

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Generate some random data for demonstration
np.random.seed(111)
data = [[1,3],[3,3],[4,3],[5,3],[1,2],[4,2],[1,1],[2,1]]

# Perform hierarchical clustering
linkage_matrix = linkage(data, method='single')
print(linkage_matrix)

# Specify the number of clusters you want
num_clusters = 1

# Assign data points to clusters using fcluster
clusters = fcluster(linkage_matrix, t=num_clusters, criterion='maxclust')

# Plot the dendrogram with color-coded clusters
dendrogram(linkage_matrix, color_threshold=num_clusters, labels=np.arange(1, 9))
# dendrogram(linkage_matrix, color_threshold=num_clusters)
plt.title('Hierarchical Clustering Dendrogram with {} Clusters'.format(num_clusters))
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Display the assigned clusters for each data point
print("Assigned clusters:", clusters)
```
